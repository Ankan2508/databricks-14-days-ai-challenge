```python
%sql
-- Create structure
CREATE CATALOG ecommerce;
USE CATALOG ecommerce;
CREATE SCHEMA bronze;
CREATE SCHEMA silver;
CREATE SCHEMA gold;
```

```python
bronze_df = spark.read.format("delta").load("/Volumes/workspace/ecommerce/ecommerce_data/bronze_events")
bronze_df.createOrReplaceTempView("bronze_events")

silver_df = spark.read.format("delta").load("/Volumes/workspace/ecommerce/ecommerce_data/silver_events")
silver_df.createOrReplaceTempView("silver_events")

gold_df = spark.read.format("delta").load("/Volumes/workspace/ecommerce/ecommerce_data/gold_events")
gold_df.createOrReplaceTempView("gold_events")
```

```python
%sql
SELECT * FROM bronze_events;
SELECT * FROM silver_events;
SELECT * FROM gold_events;
```

```python
gold_df.write.format("delta").mode("overwrite").saveAsTable("ecommerce.gold.products")
```
