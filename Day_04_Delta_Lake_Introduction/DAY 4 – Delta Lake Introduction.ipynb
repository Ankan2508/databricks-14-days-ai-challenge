```python
events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv", header=True, inferSchema=True)
```

```python
# Convert to Delta
events.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.events_table")
```

```python
display(
    spark.sql(
        """
        DESCRIBE DETAIL workspace.default.events_table
        """
    )
)
```

```python
# Test schema enforcement
try:
    wrong_schema = spark.createDataFrame(
        [("a","b","c")], 
        ["x","y","z"]
        )
    
    wrong_schema.write \
        .format("delta") \
        .mode("append") \
        .save("/delta/events")

except Exception as e:
    print(f"Schema enforcement: {e}")
```
